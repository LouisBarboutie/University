```c++
#include <stdio.h>     
#include <stdlib.h>    
#include <string.h>    
#include <math.h>   

void pascal(int N)
{
    int *triangle;
    int i,j;
    
    // get the N*N Array, it's a waste of memory, but makes it easier to get the job done
    triangle = (int *)malloc(N*N*sizeof(int));
    
    // fill the Array with 0
    for ( i = 0; i < N; i++)
    {
        for ( j = 0; j < N; j++)
        {
            triangle[i * N + j] = 0;
        }
    }
    
    // Build Pascal Triangle
    triangle[0]   = 1;

    for ( i = 1; i < N; i++ )
    {
        for ( j = 0; j < N; j++)
        {
            triangle[i*N+j] = triangle[(i-1)*N+j]+triangle[(i-1)*N+j-1];
        }
    }
    
    // janky print
    for ( i = 0; i < N; i++)
    {
        for ( int k = 0; k < (N-i); k++)
        {
            printf("   ");
        }
        
        for ( j = 0; j < N; j++)
        {
            // print the entries
            if (triangle[i * N + j] != 0)
            {
                printf("%*i",6,triangle[i * N + j]);
            }
            // don't print 0
            else { 
                //printf("%i ",triangle[i * N + j]);
                printf("");
            }
        }
        
        printf("\n");
    }
    
    free(triangle);
}

pascal(15)
```
